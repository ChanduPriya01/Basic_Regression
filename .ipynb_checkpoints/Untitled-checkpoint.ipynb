# Loading the data and the required packages

```python
# importing the packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import preprocessing
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# importing the data
data_train = pd.read_csv(r"C:\Users\chand\Desktop\Qlaym_DataScientist_Assignment\data\dataset_known.csv")
data_test = pd.read_csv(r"C:\Users\chand\Desktop\Qlaym_DataScientist_Assignment\data\dataset_unknown.csv")
```

# Preprocessing

```python
data_train.describe()
```

```python
# checking the missing values
for i in data_train.columns:
    print("{} -{}".format(i, data_train[i].isna().sum().sum()))
print("Training data has a total missing values of", data_train.isna().sum().sum())
```

```python
def check_missing_values():
    print("Length of training data is", len(data_train))
    data_no_missing_values = data_train.dropna(axis=0)
    print("Length of training data after removing the missing values is", len(data_no_missing_values))
```

```python
# checking the no.of numerical values
numerical_columns = []
categorical_columns = []
for i in data_train.columns:
    if data_train[i].dtype != "object" :
        numerical_columns.append(i)
    else:
        categorical_columns.append(i)
print(len(numerical_columns))
```

```python
# filling the missing values of numerical columns with the mean of that column
for i in numerical_columns:
    data_train[i].fillna(data_train[i].mean(), inplace=True)
```

```python
check_missing_values()
```

```python
for i in categorical_columns:
    print(data_train[i].unique())
    data_check = data_train.replace("<undefined>", np.NaN, inplace=True)
```

```python
check_missing_values()
```

Since dropping the "undefined" value does not affect the data much, the rows are dropped

```python
data_train = data_train.dropna(axis=0)
```

# Data Exploration

```python
# checking for duplicate columns with different name
remove = []
cols = data_train.columns
for i in range(len(cols)-1):
    v = data_train[cols[i]].values
    for j in range(i+1,len(cols)):
        if np.array_equal(v,data_train[cols[j]].values):
            remove.append(cols[j])
print("The number of duplicate columns are", len(remove))
```

```python
target = data_train['target']
```

```python
sns.boxplot(x=target)
```

It can be seen that there are many outliers in the target variable.

```python
data_input = data_train.drop(["target"], axis = 1)
```

```python
data_input
```

```python
# scaling the numerical values to a scale of (0,1)
min_max_scaler = preprocessing.MinMaxScaler() 
numerical_columns.remove("target")
for i in numerical_columns:
    data_input[i] = min_max_scaler.fit_transform(np.array(data_input[i]).reshape(-1, 1))
```

```python
# one hot coding the categorical columns
data_input_one_hot = data_input
for i in categorical_columns:
    data_input_one_hot = pd.concat([data_input_one_hot, pd.get_dummies(data_input[i], prefix=i)], axis=1)
    data_input_one_hot.drop([i], axis=1, inplace=True)
```

```python
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(data_input_one_hot, target)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(data_input.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```
